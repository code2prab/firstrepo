```python
# 1st cluster
array1=np.random.randn(100,2)
x1=array1[:, 0]+0
y1=array1[:, 1]+0
centroid_1 = array1.mean(axis=0)
plt.scatter(x1,y1,color="blue",marker='v')
plt.scatter(centroid_1[0],centroid_1[1],s =300,c='black',marker ='P')

# 2nd cluster
array2=np.random.randn(100,2)
x2=array2[:,0]+0
y2=array2[:,1]+2
centroid_2 = array2.mean(axis=0)
plt.scatter(x2,y2,color='red',marker='o')
plt.scatter(centroid_2[0],centroid_2[1],s =300,c='black',marker ='P')

# 3rd cluster
array3 = np.random.randn(100,2)
x3 = array3[:,0]+1
y3 = array3[:,1]+1
centroid_3 = array3.mean(axis=0)
plt.scatter(x3,y3,color='yellow',marker='s')
plt.scatter(centroid_3[0],centroid_3[1],s =300,c='black',marker ='P')


# 4th cluster
array4 = np.random.randn(100,2)
x4=array4[:,0]+2
y4=array4[:,1]+0
centroid_4 = array4.mean(axis=0)
plt.scatter(x4,y4,color='cyan',marker='p')
plt.scatter(centroid_4[0],centroid_4[1],s =300,c='black',marker ='P')


# 5th cluster
array5 = np.random.randn(100,2)
x5 = array5[:, 0]+2
y5 = array5[:, 1]+2
centroid_5 = array5.mean(axis=0)
plt.scatter(x5,y5,color='green',marker='*')
plt.scatter(centroid_5[0],centroid_5[1],s =300,c='black',marker ='P')
```
